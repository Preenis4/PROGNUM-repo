```python
import math
from scipy.constants import G

def v_esc(mass, radius):
    res = math.sqrt((2*G*mass)/radius)
    return res//1

masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']

for m,r,n in zip(masses,radii,names):
    print(f"V_esc = {v_esc(m, r*1000)}. Mass = {m:.2f}. Name: {n}.")
```

```python
import math
import numpy as np

def mean(arr):
    sum = 0
    for i in arr:
        sum += i
    return sum/len(arr)

def stdev(arry):
    mid = 0
    meann = mean(arry)
    for j in arry:
        mid += (j - meann)**2
    return math.sqrt(mid/len(arry))

myarr = list(np.random.normal(2,3, 100))
print(mean(myarr))
print(stdev(myarr))
print(np.mean(myarr))
print(np.std(myarr))
```

```python
x = [18.49, 35.48, 35.48, 11.34, 7.12, 9.82, 23.66, 44, 66.7, 52.4, 55, 25.6, 26.19]
y = [2094.75, 2320.87, 1224.9, 872.13, 432.2, 751, 1410, 2700, 5292, 2550, 5253, 2419, 2158]
x_avg = np.sum(x)/len(x)
y_avg = np.sum(y)/len(y)

x = np.array(x)
y = np.array(y)


def fun(x, y):
    top = np.sum((x-x_avg)*(y-y_avg))
    bot = np.sum((x-x_avg)**2)
    b = top/bot
    a = y_avg - (b*x_avg)
    return float(a), float(b)


print(fun(x, y))
```
